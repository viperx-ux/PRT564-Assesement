# NSW Road Crash Analysis (2020–2024)

PRT564 Assessment 2 – Group Project Presentation  
Darwin Group 12

## Project Overview
This project analyses NSW road crash data (2020–2024) to identify geographical and temporal patterns, and to evaluate relationships between crash factors and total casualties.

## Research Questions
1. How do location factors affect crash occurrence and severity?  
2. How does crash frequency change over time?  
3. How do environmental conditions influence crashes? *(future work)*  
4. Is there a relationship between traffic units and total casualties?  

## Files
- **preprocessing.py** – initial data preprocessing  
- **preprocessing_eda.py** – preprocessing, feature engineering, and EDA (generates figures)  
- **Simple Linear Regression.py** – simple linear regression model  
- **Multiple Linear Regression.py** – multiple linear regression model  

## Data
- **nsw_road_crash_data_2020-2024_crash.xlsx** – original dataset  
- **nsw_crash_preprocessed.csv** – cleaned dataset used for analysis and modelling  

## Output Figures
Figures generated from EDA and regression analysis are included in this repository (e.g., severity distribution, temporal trends, geographical patterns, regression diagnostics).
